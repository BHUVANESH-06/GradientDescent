```python
import numpy as np 
import matplotlib.pyplot as plt 

# Generating 1000 points between 0 and 10
X = np.linspace(0, 10, 1000)
print(len(X)) 

# Generating Y values using the equation Y = 3X + 4 with some noise using normal distribution

Y = 3 * X + 4 + np.random.normal(0, 1)

# Initializing the slope (m1) randomly and continuously changing it according to the gradient

m1 = np.random.rand()

# Lists to store slope values, loss values, and predicted Y values

m_values = []
loss = []
y_predval = []

# Function to perform gradient descent

def fun(X, Y):
    global m1
    N = len(X)  # Number of data points
    
    best_m = float('inf')  # Initializing best loss as a high value for best m value
    
    epoch = 0  # Variable to store the best epoch
    
    best_y_pred = []  # To store the best prediction

    # Running gradient descent for 1000 iterations
    
    for iteri in range(1000):
        # Predicted Y using current m1
        Y_pred = m1 * X + 4  
        y_predval.append(Y_pred)

        # Calculating Mean Squared Error (Loss)
        current_loss = np.mean((Y - Y_pred) ** 2)
        loss.append(current_loss)  

        # Checking if the current loss is the best so far
        
        if current_loss < best_m:
            epoch = iteri  # Storing the epoch where best loss is found
            best_y_pred = Y_pred  # Storing the best prediction
            best_m = current_loss  # Updating best loss

        # Computing gradient of loss w.r.t. m1 which is got by differentiting the equation ((y-y')^2/N) where y'=m1x+m2 w.r.t m1;
        
        grad_m = (-2 / N) * np.sum(X * (Y - Y_pred))

        # Storing current slope value
        m_values.append(m1)

        # Updating m1 using gradient descent (learning rate = 0.001)
        m1 = m1 - 0.001 * grad_m

        if iteri % 1000 == 0:
            pass

    return m1, epoch, best_y_pred

# Calling the function to perform gradient descent
m1, epoch, best_y_pred = fun(X, Y)

# Printing final slope and epoch where it was found
print("Predicted M1 Value:", m1)
print("Epoch I got this m is:", epoch)

# Plotting actual vs predicted data
plt.plot(X, Y, label="Actual Data", alpha=0.6)  
plt.plot(X, best_y_pred, label="Predicted Line", color='r', linewidth=2) 
plt.legend()
plt.title("Actual vs Predicted")
plt.show()

# Plotting loss vs slope during gradient descent
plt.plot(m_values, loss)
plt.xlabel("Slope (m1)")
plt.ylabel("Loss")
plt.title("Loss vs Slope during Gradient Descent")
plt.show()
```
